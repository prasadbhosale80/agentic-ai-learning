# Batch

## What is Batch?

**Batch** = send multiple questions at once instead of one by one.

Without batch → call `.invoke()` 5 times one after another (slow)  
With batch → send all 5 together and get all 5 answers at once (fast!)

| | `.invoke()` | `.batch()` |
|---|---|---|
| Input | One question (string) | List of questions |
| Output | One response | List of responses |
| Speed for many questions | Slower | Faster |

> **Real-world use cases:** summarize 100 articles, translate many sentences, classify support tickets
>
> **Note:** Order is preserved — `responses[0]` answers your first question, `responses[1]` answers the second, etc.

```python
from langchain_groq import ChatGroq

model = ChatGroq(
    model = "llama-3.1-8b-instant",
    temperature = 0.7
)

responses = model.batch(["Who build Taj Mahal, give me just person name", "how far moon from earth"])
for response in responses:
    print(response.content)
```
