# Streaming


## What is Streaming?

Normally `.invoke()` waits for the **full answer** to load — like waiting for a complete download.

**Streaming** shows the answer **word by word** as it's generated — just like how ChatGPT types out the reply in real time.

| | `.invoke()` | `.stream()` |
|---|---|---|
| Output | All at once | Word by word |
| User experience | Wait then see | See as it types |
| Good for | Short answers, background tasks | Long answers, live chat UI |

> **Code tips for the cell below:**
> - `end=""` → don't add a new line after each chunk (keeps text on the same line)
> - `flush=True` → print immediately without buffering
> - `chunk.text` and `chunk.content` both work
> - `temperature=0.7` → controls creativity (0 = same every time, 1 = more random)

```python
from langchain_groq import ChatGroq

model = ChatGroq(
    model = "llama-3.1-8b-instant",
    temperature = 0.7
)

response = model.stream("Tell me history of taj mahal in 500 words")
for chunk in response:
    print(chunk.text, end= "", flush=True)
    #print(chunk.content, end= "", flush=True)
```
