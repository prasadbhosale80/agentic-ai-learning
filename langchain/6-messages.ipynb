## Types of Messages

In LangChain, a **conversation is just a list of messages**. Each message has a type that tells the AI who said it and what role it plays.

### 1. `SystemMessage`
- **Created by:** Developer (you)
- **Purpose:** Sets the AI's personality, role, or rules. The user never sees this.
- **Example:** `SystemMessage("You are a helpful assistant who speaks simply.")`
- Think of it as **backstage instructions** before the show starts.

### 2. `HumanMessage`
- **Created by:** The end user
- **Purpose:** Represents what the human typed or asked.
- **Example:** `HumanMessage("What is the weather in Pune?")`

### 3. `AIMessage`
- **Created by:** The AI model
- **Purpose:** Represents what the AI responded.
- **Example:** `AIMessage("Weather in Pune is 22°C")`

### 4. `ToolMessage`
- **Created by:** A tool/function
- **Purpose:** Carries the result returned by a tool back to the agent.
- **Example:** `ToolMessage("22 deg C")`
- Think of it as the **tool handing its answer back** so the AI can form a reply.

### Message Flow
```
[SystemMessage]  ← Sets rules (hidden from user)
[HumanMessage]   ← User asks a question
[ToolMessage]    ← Tool returns data
[AIMessage]      ← AI forms final reply
```
